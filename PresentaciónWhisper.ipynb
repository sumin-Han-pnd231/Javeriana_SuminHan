```python
import whisper

model = whisper.load_model("turbo")
result = model.transcribe("/content/mark.wav")
print(result["text"])
```

```python
#Ejemplo transcripción No ingles
result2 = model.transcribe("/content/kor.mp3")
print(result2["text"])
```

```python
#Ejemplo Traducción No ingles a Ingles
model2 = whisper.load_model("medium")
result3 = model2.transcribe("/content/ejspa.mp3", task='translate')
print(result3["text"])
```
